```python
import pdb
# pdb.set_trace()
def findchar(ori,char):
    for col in range(ori.shape[1]-char.shape[1]):
        for row in range(ori.shape[0]-char.shape[0]):
#             print((row,col))
#             print(ori[row:row+char.shape[0],col:col+char.shape[1]].shape)
            if (ori[row:row+char.shape[0],col:col+char.shape[1]]==char).all():
                return (row,col)
    return None
            
```
